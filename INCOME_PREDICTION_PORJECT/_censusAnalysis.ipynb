```python
def WorkClass(pos):
    if pos['workclass'] == "Private":
        return "Private"
    elif pos['workclass'] == 'State-gov' or pos['workclass']== "Federal-gov" or pos['workclass'] == "Local-gov":
        return "Government"
    elif pos['workclass']== "Self-emp-not-inc" or pos['workclass'] == "Self-emp-inc":
        return "Self_emp"
    else:
        return "Without_ pay"
df['workclass'] = df.apply(WorkClass,axis = 1)
```

```python
# 1. couple - 'Married-civ-spouse','Married-AF-spouse'
# 2. single - 'Never-married', 'Divorced', 'Separated', 'Widowed', 'Married-spouse-absent',
```

```python
def maritalstatus(status):
    if status['marital-status'] == 'Married-civ-spouse' or status['marital-status'] == 'Married-AF-spouse':
        return 'couple'
    else:
        return 'single'
df['marital-status'] = df.apply(maritalstatus,axis=1)
```

```python
def countrycode(cd):
  if cd['native-country'] == 'United-States':
    return 'US'
  else:
    return 'Others'
df['native-country'] = df.apply(countrycode,axis=1)
```

```python
def Education_level(EL):
  if EL['education'] == 'Preschool' or EL['education'] == '1st-4th' or EL['education'] == '5th-6th':
    return "primary"
  elif EL['education'] == '7th-8th':
    return "Middle"
  elif EL['education'] == '9th' or EL['education'] == '10th' or EL['education'] == '11th' or EL['education'] == '12th':
    return "HighSchool"
  elif EL['education'] == 'HS-grad' or EL['education'] == 'Some-college' or EL['education'] == 'Assoc-acdm' or EL['education'] == 'Assoc-voc':
    return "highersecondary"
  elif EL['education'] == 'Bachelors':
    return "Bachelors"
  elif EL['education'] == 'Masters':
    return "Masters"
  else:
    return "Doctrate"

df['education'] = df.apply(Education_level, axis=1)
```

```python
def income(salary):
  if salary['income'] == '<=50K' or salary['income'] == "<=50K.":
    return "<=50k"
  else:
    return ">50k"
df['income'] = df.apply(income,axis=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
df.info()
```

```python
numerical_features = ['age', 'fnlwgt', 'education-num',
                      'capital-gain', 'capital-loss',
                      'hours-per-week']
```

```python
categorical_features = ['workclass', 'education', 'marital-status',
                        'occupation', 'relationship', 'race',
                        'sex', 'native-country']
```

# Preprocessing pipeline

```python
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

```python
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
processor = ColumnTransformer(
    transformers=[
        ('num',numerical_transformer,numerical_features),
        ('cat',categorical_transformer,categorical_features)
    ]
)
```

```python
df['income'].info()
```

```python
processor
```

```python
pipeline = Pipeline(steps=
         [('preprocess', processor),
          ('model', RandomForestClassifier())
])
```

```python
pipeline
```

```python
param_grid = {
    'model__n_estimators': [50, 100, 200],
    'model__max_depth': [5, 10, 15],
    'model__criterion': ['gini', 'entropy']
}
```

```python
# Feature selection
X = df.drop('income',axis=1)
y = df['income']
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42,stratify=y)
```

```python
grid = GridSearchCV(estimator=pipeline,
                    param_grid=param_grid,
                    cv=3,
                    scoring='accuracy'
                    )
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
best_model = grid.best_estimator_
best_model
```

```python
y_pred = grid.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred)*100)
print(confusion_matrix(y_test, y_pred))
```
